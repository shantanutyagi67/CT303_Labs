```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import pandas as pd
import math
```

```python
def part_a(A,T,n):
    fs = 10/T
    ts = 1/fs
    x = np.empty([n, int(T/ts)])
    for i in range(n):
        num = np.random.randint(2)
        if num == 0:
            num = -A
        else:
            num = A
        row = np.full((1, int(T/ts)), num)
        x[i] = row;
    
    y_axis = np.reshape(x,(n*int(T/ts),1))
    x_axis = np.arange(0,n,ts/T)
    print(np.shape(x_axis))
    
    sigma = 1;
    awgn_sequence = sigma*(np.random.randn(int(n*T*fs)));
    print(np.shape(awgn_sequence))
    
    #plotting
    plt.figure()
    plt.plot(x_axis,y_axis)
    plt.title('NRZL Scheme')
    plt.show()
    
    transmitted_signal = awgn_sequence;
    for i in range(int(n*T*fs)):
        transmitted_signal[i] = transmitted_signal[i] + y_axis[i]
    print(np.shape(transmitted_signal))
    plt.figure()
    plt.plot(x_axis,transmitted_signal)
    plt.title('Transmitted Signal')
    plt.show()
    return transmitted_signal
```

```python
T=0.2
n=16
A = 2;
experiments = 1000;
transmitted_signal = part_a(A,T,n)
```
