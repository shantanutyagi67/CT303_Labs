```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q1(a) X(n) = cos(0.2πn + θ), θ ∼ U[−π, π], n ∈ [0, 9]

```python
def f1(N):
    sz = int(10);
    theta = np.zeros((sz,N))
    n = np.zeros((sz,N))
    lis = np.random.uniform(-np.pi,np.pi,N)
    for i in range(sz):
        n[i] = i
        theta[i] = lis
    y = np.cos(0.2*np.pi*n + theta)
    mean_y = np.mean(y,axis = 1);
    
    plt.stem(mean_y)
    plt.axhline(y=np.mean(mean_y), color='r', linestyle='-')
    plt.title('Mean')
    plt.xlabel('n')
    plt.ylabel('X(n)')
    plt.show()
    autocorrelation = np.zeros((sz,sz))
    for i in range(sz):
        for j in range(sz):
            autocorrelation[i][j] = np.correlate(y[i],y[j])
    print("Correlation matrix:\n",autocorrelation,"\n")
    x1 = np.correlate(y[3], y[1])
    x2 = np.correlate(y[4], y[2])
    x3 = np.correlate(y[5], y[3])
    x4 = np.correlate(y[6], y[4])
    x5 = np.correlate(y[7], y[5])
    x6 = np.correlate(y[8], y[6])
    x7 = np.correlate(y[9], y[7])
    print("correlation of y3 with y1",x1)
    print("correlation of y4 with y2",x2)
    print("correlation of y5 with y3",x3)
    print("correlation of y6 with y4",x4)
    print("correlation of y7 with y5",x5)
    print("correlation of y8 with y6",x6)
    print("correlation of y9 with y7",x7)
    
    #Histogram
    bins=25
    plt.hist(y[2],bins, facecolor='red', alpha=0.6)
    plt.title(r'Histogram at index 2')
    plt.xlabel('X(n)')
    plt.ylabel('Number of realizations')
    plt.show()
    plt.hist(y[7],bins, facecolor='red', alpha=0.6)
    plt.title(r'Histogram at index 7')
    plt.xlabel('X(n)')
    plt.ylabel('Number of realizations')
    plt.show()
```

```python
f1(10000)
```

```python
#We see that the mean is 0 and the autocorrelation depends on the difference of n1 and n2,
#therefore this equation follows wide-sense stationarity
```

```python
f3(10000)
```

```python
#We see that the mean is 0 and the autocorrelation depends on the difference of n1 and n2,
#therefore this equation follows wide-sense stationarity
```

### loading and plotting the data

```python
import pandas as pd
df = pd.read_excel (r'Gandhinagar_RainfallData.xls')
df
```

```python
import numpy as np
array = df.to_numpy()
data = array[1:,1:]
data
data.mean()
value = data.reshape(1,102*12)
import matplotlib.pyplot as plt 
plt.figure(figsize=(20,10))
plt.plot( np.linspace(1,102*12,102*12),value[0]) 
  
# naming the x axis 
plt.xlabel('time') 
# naming the y axis 
plt.ylabel('rainfall value') 
  
# giving a title to my graph 
plt.title('Gandhinagar Rainfall') 
  
# function to show t
plt.show() 
```

```python
# since the rainfall data has seasonal trends it is logically a non stationary process,
# we can see this by plotting the graph for a given year as
```

### Overall mean

```python
data.mean()
```

### Mean monthly data 

```python
monthly = np.array([])
for i in range(12):
    monthly = np.append(monthly,data[0:][i].mean())
monthly
```

```python
plt.figure(figsize=(20,10))
plt.plot( np.linspace(1,12,12),monthly) 
  
# naming the x axis 
plt.xlabel('month') 
# naming the y axis 
plt.ylabel('rainfall value') 
  
# giving a title to my graph 
plt.title('Gandhinagar Rainfall') 
  
# function to show t
plt.show() 
```

### Mean yearly data

```python
yearly = np.array([])
for i in range(102):
    yearly = np.append(yearly,data[0:][i].mean())
yearly
```

```python
plt.figure(figsize=(20,10))
plt.plot( np.linspace(1,102,102),yearly) 
  
# naming the x axis 
plt.xlabel('year') 
# naming the y axis 
plt.ylabel('rainfall value') 
  
# giving a title to my graph 
plt.title('Gandhinagar Rainfall') 
  
# function to show t
plt.show() 
```

```python
# from the monthly mean we can conclue that the mean varies largely with the month
# from the yearly mean we can conclude that average rainfall every year varies slightly around the mean
```

### Covariance Heatmap

```python
import seaborn as sn
covMatrix=np.cov(data.astype(float), rowvar=False)
plt.subplots(figsize=(20,20)) 
sn.heatmap(covMatrix, annot=True, fmt='g', vmin=-1, vmax=1, center=0, cmap='coolwarm')
plt.show()
```

```python
# thus we see no covariance and thus conclude that the stochastic process is non stationary 
# (both in wide-sense and strict-sense)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
def f1(N):
    sz = int(10);
    theta = np.zeros((sz,N))
    n = np.zeros((sz,N))
    lis = np.random.uniform(-np.pi,np.pi,N)
    for i in range(sz):
        n[i] = i
        theta[i] = lis
    y = np.cos(0.2*np.pi*n + theta)
    mean_y = np.mean(y,axis = 1);
    plt.stem(mean_y)
    plt.axhline(y=np.mean(mean_y), color='r', linestyle='-')
    plt.show()
    autocorrelation = np.zeros((sz,sz))
    for i in range(sz):
        for j in range(sz):
            autocorrelation[i][j] = np.correlate(y[i],y[j])
    
    print("Correlation Matrix, R = ", autocorrelation)
    
    transpose_mat = np.transpose(autocorrelation)    
    eigen_mat1 = np.linalg.eigvalsh(autocorrelation)
    
    if np.all(eigen_mat1 > -(1e-7)) and np.all(np.abs(autocorrelation - transpose_mat) < 1e-7):
        print("autocorrelation matrix is positive semi-definite\n")
    else:
        print("autocorrelation matrix is not positive semi-definite\n")
    
    y_minus_m = y
    for i in range(sz):
        y_minus_m[i] = y_minus_m[i] - mean_y[i]
        
    autocovariance = np.zeros((sz,sz))
    for i in range(sz):
        for j in range(sz):
            autocovariance[i][j] = np.correlate(y_minus_m[i],y_minus_m[j])
    
    print("covariance matrix, C: ", autocovariance)
    
    transpose_mat = np.transpose(autocovariance)
    eigen_mat = np.linalg.eigvalsh(autocovariance)
    
    if np.all(eigen_mat1 > -(1e-7)) and np.all(np.abs(autocovariance - transpose_mat) < 1e-7):
        print("autocovariance matrix is positive semi-definite\n")
    else:
        print("autocovariance matrix is not positive semi-definite\n")
    
    plt.figure(figsize=(10, 8))
    plt.title(r'Heatmap of correlation')
    sn.heatmap(autocorrelation, annot=True, cmap='coolwarm')
    
    plt.figure(figsize=(10, 8))
    plt.title(r'Heatmap of covariance')
    sn.heatmap(autocovariance, annot=True, cmap='coolwarm')
```

```python
f1(10000)
```
