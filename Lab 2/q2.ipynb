### loading and plotting the data

```python
import pandas as pd
df = pd.read_excel (r'Gandhinagar_RainfallData.xls')
df
```

```python
import numpy as np
array = df.to_numpy()
data = array[1:,1:]
data
data.mean()
value = data.reshape(1,102*12)
import matplotlib.pyplot as plt 
plt.figure(figsize=(20,10))
plt.plot( np.linspace(1,102*12,102*12),value[0]) 
  
# naming the x axis 
plt.xlabel('time') 
# naming the y axis 
plt.ylabel('rainfall value') 
  
# giving a title to my graph 
plt.title('Gandhinagar Rainfall') 
  
# function to show t
plt.show() 
```

```python
# since the rainfall data has seasonal trends it is logically a non stationary process,
# we can see this by plotting the graph for a given year as
```

### Average monthly data

```python
monthly = np.array([])
for i in range(12):
    monthly = np.append(monthly,data[0:][i].mean())
monthly
```

```python
plt.figure(figsize=(20,10))
plt.plot( np.linspace(1,12,12),monthly) 
  
# naming the x axis 
plt.xlabel('month') 
# naming the y axis 
plt.ylabel('rainfall value') 
  
# giving a title to my graph 
plt.title('Gandhinagar Rainfall') 
  
# function to show t
plt.show() 
```

### Average yearly data

```python
yearly = np.array([])
for i in range(102):
    yearly = np.append(yearly,data[0:][i].mean())
yearly
```

```python
plt.figure(figsize=(20,10))
plt.plot( np.linspace(1,102,102),yearly) 
  
# naming the x axis 
plt.xlabel('year') 
# naming the y axis 
plt.ylabel('rainfall value') 
  
# giving a title to my graph 
plt.title('Gandhinagar Rainfall') 
  
# function to show t
plt.show() 
```

```python
# from the monthly mean we can conclue that the mean varies largely with the month
# from the yearly mean we can conclude that average rainfall every year varies slightly around the mean
```

```python
import seaborn as sn
covMatrix=np.cov(data.astype(float), rowvar=False)
plt.subplots(figsize=(20,20)) 
sn.heatmap(covMatrix, annot=True, fmt='g', vmin=-1, vmax=1, center=0, cmap='coolwarm')
plt.show()
```

```python
# thus we see no covariance and thus conclude that the stochastic process is non stationary 
# (both in wide-sense and strict-sense)
```
