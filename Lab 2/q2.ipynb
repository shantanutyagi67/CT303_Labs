```python
value = data.reshape(1,102*12)
```

```python
value[0]
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(20,10))
plt.plot( np.linspace(1,102*12,102*12),value[0]) 
  
# naming the x axis 
plt.xlabel('time') 
# naming the y axis 
plt.ylabel('rainfall value') 
  
# giving a title to my graph 
plt.title('Gandhinagar Rainfall') 
  
# function to show t
plt.show() 
```
