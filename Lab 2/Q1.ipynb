```python
import numpy as np
import matplotlib.pyplot as plt
```

# Q1(a) X(n) = cos(0.2πn + θ), θ ∼ U[−π, π], n ∈ [0, 9]

```python
def f1(N):
    sz = int(10);
    theta = np.zeros((sz,N))
    n = np.zeros((sz,N))
    lis = np.random.uniform(-np.pi,np.pi,N)
    for i in range(sz):
        n[i] = i
        theta[i] = lis
    y = np.cos(0.2*np.pi*n + theta)
    mean_y = np.mean(y,axis = 1);
    
    plt.stem(mean_y)
    plt.axhline(y=np.mean(mean_y), color='r', linestyle='-')
    plt.title('Mean')
    plt.xlabel('n')
    plt.ylabel('X(n)')
    plt.show()
    autocorrelation = np.zeros((sz,sz))
    for i in range(sz):
        for j in range(sz):
            autocorrelation[i][j] = np.correlate(y[i],y[j])
    print("Correlation matrix:\n",autocorrelation,"\n")
    x1 = np.correlate(y[3], y[1])
    x2 = np.correlate(y[4], y[2])
    x3 = np.correlate(y[5], y[3])
    x4 = np.correlate(y[6], y[4])
    x5 = np.correlate(y[7], y[5])
    x6 = np.correlate(y[8], y[6])
    x7 = np.correlate(y[9], y[7])
    print("correlation of y3 with y1",x1)
    print("correlation of y4 with y2",x2)
    print("correlation of y5 with y3",x3)
    print("correlation of y6 with y4",x4)
    print("correlation of y7 with y5",x5)
    print("correlation of y8 with y6",x6)
    print("correlation of y9 with y7",x7)
    
    #Histogram
    bins=25
    plt.hist(y[2],bins, facecolor='red', alpha=0.6)
    plt.title(r'Histogram at index 2')
    plt.xlabel('X(n)')
    plt.ylabel('Number of realizations')
    plt.show()
    plt.hist(y[7],bins, facecolor='red', alpha=0.6)
    plt.title(r'Histogram at index 7')
    plt.xlabel('X(n)')
    plt.ylabel('Number of realizations')
    plt.show()
```

```python
f1(10000)
```

```python
#We see that the mean is 0 and the autocorrelation depends on the difference of n1 and n2,
#therefore this equation follows wide-sense stationarity
```

# Q1(b) X(t) = A cos(0.25πn), A ∼ U[−5, 5], n ∈ [0, 7]

```python
def f2(N):
    sz = int(8);
    A = np.zeros((sz,N))
    n = np.zeros((sz,N))
    lis = np.random.uniform(-5,5,N)
    for i in range(sz):
        n[i] = i
        A[i] = lis
    y = A* np.cos(0.25*np.pi*n)
    mean_y = np.mean(y,axis = 1);
    plt.stem(mean_y)
    plt.title('Mean')
    plt.xlabel('n')
    plt.ylabel('X(n)')
    plt.axhline(y=np.mean(mean_y), color='r', linestyle='-')
    plt.show()
    
    autocorrelation = np.zeros((sz,sz))
    for i in range(sz):
        for j in range(sz):
            autocorrelation[i][j] = np.correlate(y[i],y[j])
    print("Correlation matrix:\n",autocorrelation,"\n")
    x1 = np.correlate(y[3], y[1])
    x2 = np.correlate(y[4], y[2])
    x3 = np.correlate(y[5], y[3])
    x4 = np.correlate(y[6], y[4])
    x5 = np.correlate(y[7], y[5])
    print("correlation of y3 with y1",x1)
    print("correlation of y4 with y2",x2)
    print("correlation of y5 with y3",x3)
    print("correlation of y6 with y4",x4)
    print("correlation of y7 with y5",x5)
    
    #Histogram
    bins=25
    plt.hist(y[2],bins, facecolor='red', alpha=0.6)
    plt.title(r'Histogram at index 2')
    plt.xlabel('X(n)')
    plt.ylabel('Number of realizations')
    plt.show()
    plt.hist(y[7],bins, facecolor='red', alpha=0.6)
    plt.title(r'Histogram at index 7')
    plt.xlabel('X(n)')
    plt.ylabel('Number of realizations')
    plt.show()
```

```python
f2(10000)
```

```python
#We see that the mean is 0 but the autocorrelation does not depend on the difference of n1 and n2,
#therefore this equation does not follows wide-sense stationarity
```

# Q1(c) X(n) = A(n), where A(n) ∼ N (0, 1) 

```python
def f3(N):
    sz = int(10);
    lis = []
    y = []
    for i in range(0,N):
        np.random.seed(i+1)
        lis.append(np.random.normal(0,1,1000))
    mean_y = []
    for i in range(0,1000,100):
        temp=[]
        for j in range(0,N):
            temp.append(lis[j][i])
        y.append(temp);
        mean_y.append(np.mean(temp))
    
    plt.stem(mean_y)
    plt.title('Mean')
    plt.xlabel('n')
    plt.ylabel('X(n)')
    plt.show()
    
    autocorrelation = np.zeros((sz,sz))
    for i in range(sz):
        for j in range(sz):
            autocorrelation[i][j] = np.correlate(y[i],y[j])
    print("Correlation matrix:\n",autocorrelation,"\n")
    x1 = np.correlate(y[3], y[1])
    x2 = np.correlate(y[4], y[2])
    x3 = np.correlate(y[5], y[3])
    x4 = np.correlate(y[6], y[4])
    x5 = np.correlate(y[7], y[5])
    print("correlation of y3 with y1",x1)
    print("correlation of y4 with y2",x2)
    print("correlation of y5 with y3",x3)
    print("correlation of y6 with y4",x4)
    print("correlation of y7 with y5",x5)
    
    #Histogram
    bins=25
    plt.hist(y[2],bins, facecolor='red', alpha=0.6)
    plt.title(r'Histogram at index 2')
    plt.xlabel('X(n)')
    plt.ylabel('Number of realizations')
    plt.show()
    plt.hist(y[7],bins, facecolor='red', alpha=0.6)
    plt.title(r'Histogram at index 7')
    plt.xlabel('X(n)')
    plt.ylabel('Number of realizations')
    plt.show()
```

```python
f3(10000)
```

```python
#We see that the mean is 0 and the autocorrelation depends on the difference of n1 and n2,
#therefore this equation follows wide-sense stationarity
```
