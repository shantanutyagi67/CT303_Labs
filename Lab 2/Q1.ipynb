## (a)

```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
def f1(N):
    theta = np.random.uniform(-np.pi,np.pi,N)
    n = np.random.random_sample(N)
    n = n*9
    y = np.cos(0.2*np.pi*n + theta)
    print('mean= '+ str(y.mean()))
    plt.hist(y, bins='auto')
```

```python
f1(10)
```

```python
f1(100)
```

```python
f1(1000)
```

```python
# mean is dependent of the samples taken
```

```python
def f3(N):
    mu = 0
    sigma = 1
    x = np.random.normal(mu,sigma,N)
    print('mean= '+ str(x.mean()))
    
    #plot
    num_bins = 100
    n, bins, patches = plt.hist(x, num_bins,  density = 1)  
    y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *np.exp(-0.5 * (1 / sigma * (bins - mu))**2)) 
    plt.plot(bins, y, '--', color ='red')
    plt.show() 
```
