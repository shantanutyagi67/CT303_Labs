# Question 1:

```python
import matplotlib.pyplot as plt
import numpy as np

def mysinplot(f, fs, n):
    x = np.linspace(-n/(2*f),n/(2*f),fs*n)
    plt.stem(x,np.sin(2*np.pi*f*x))
    plt.title('Sine Wave')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True, which='both')
    plt.show()

wave_frequency = int(input("Input Frequency: "))
sampling_frequency = int(input("Input Sampling Frequency: "))
no_cycles = int(input("Input number of cycles: "))
mysinplot(wave_frequency,sampling_frequency,no_cycles)
```

```python
# linspace command creates an arithmetic progression in the given interval with given terms. since we are taking n cycles
# and time duration of each cycle is -1/2f to 1/2f, we multiply these values by n to get the desired time period for n cycles.
# sampling frequency = samples/cycle = fs. so total samples in the entire interval = n*fs which is the third parameter.
# next we plot the sin wave in the form sin(2*PI*f*t) where t values are in the array named x and plot it against x (time)
```

### frequency = 2, sampling freq = 20, cycles = 1

```python
mysinplot(2, 20, 1)
```

### frequency = 2, sampling freq = 20, cycles = 1

```python
mysinplot(2, 20, 4)
```

### frequency = 1, sampling freq = 10, cycles = 4

```python
mysinplot(1, 10, 4)
```

# Question 2:

```python
# linspace command creates an arithmetic progression in the given interval with given terms. since we are taking n cycles
# and time duration of each cycle is -1/2f to 1/2f, we multiply these values by n to get the desired time period for n cycles.
# sampling frequency = samples/cycle = fs. so total samples in the entire interval = n*fs which is the third parameter.
# next we plot the sin wave in the form sin(2*PI*f*t) where t values are in the array named x and plot it against x (time)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft,fftfreq

f= 2

def myctft(T, T1, fs):
    x = np.linspace(-T,T,int(fs*f*2*T))
    y = np.linspace(-T1,T1,int(2*f*T1*fs))
    sinx = np.sin(2*np.pi*f*x)
    if T1<=T:
        siny = np.sin(2*np.pi*f*y)
    else:
        siny = sinx
        num = int((y.size - x.size)/2);
        zero_arr = np.zeros(num);
        siny = np.concatenate((zero_arr,siny), axis = 0)
        siny = np.concatenate((siny,zero_arr), axis = 0)
    
    plt.figure(1)
    plt.figure(figsize=(20,5))
    plt.plot(x,sinx,'r')
    plt.stem(y,siny)
    plt.title('Sine Wave')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True, which='both')
    plt.show()
    
    plt.figure(1)
    y_fft = 2*abs(fft(siny))/y.size
    freq = fftfreq(y.size,1/(f*fs))
    plt.plot(freq, y_fft)
    plt.title('Amplitude Spectrum of y')
    plt.xlabel('Frequency')
    plt.ylabel('Y(jw)')
    plt.grid(True, which='both')
    
T = float(input("Input T: "))
T1 = float(input("Input T1: "))
fs = float(input("Input sampling frequency: "))
myctft(T,T1,fs)
# enter 2, 1, 10 values for beter comparision
```
